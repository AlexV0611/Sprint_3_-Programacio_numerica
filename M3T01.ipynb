```python
# Exercici 3: Calcula mitjana i desviació estàndard de sepal length i petal length (cm)
# Mostrar resultat amb 2 decimals


import numpy as np


from sklearn.datasets import load_iris

iris = load_iris()

datos = iris.data

columna_sl = 0

sepal_length = [sepal_le[columna_sl] for sepal_le in datos] 

a = round(np.mean(sepal_length),2)

print("La media de 'sepal length' es", a)

desv = round(np.std(sepal_length),2)

print("La desviación estándar de 'sepal length' es:", desv)

columna_pl = 2

petal_length = [petal_le[columna_pl] for petal_le in datos]

b = round(np.mean(petal_length),2)

print("La media de 'petal length' es:", b)

desv1 = round(np.std(petal_length),2)

print("La desviación estándar de 'petal length' es:", desv1)

```

```python
# Exercici 5: La mitjana i desv. estàndard de cada característica per espècie.

import numpy as np

from sklearn.datasets import load_iris

iris = load_iris()
x = iris.data
y = iris.target

arr = np.hstack((x, y.reshape(-1, 1)))

clases = np.unique(y)

medias_por_clase = []
desv_por_clase = []

for clase in clases:
    
    datos_clase = arr[y == clase][:, :-1] 
    media_clase = np.mean(datos_clase, axis=0)
    medias_por_clase.append(media_clase)
    
    desviacion_estandar_clase = np.std(datos_clase, axis=0)
    desv_por_clase.append(desviacion_estandar_clase)

matriz_medias = np.array(medias_por_clase)

print("Matriz de medias por clase:")
print(matriz_medias)

matriz_desviaciones_estandar = np.array(desv_por_clase)

print("Matriz de Desviaciones Estándar por Clase:")
print(matriz_desviaciones_estandar)
```

```python
# Exercici 6: Calcula la quantitat de mostres per espècie

import numpy as np
from sklearn.datasets import load_iris

iris = load_iris()
clase = iris.target
datos = iris.data

especie, muestras = np.unique(clase, return_counts=True)


# creamos diccionario con zip de las 2 listas que nos interesan(target names y nº muestras)

counts_especies = dict(zip(iris.target_names, muestras))

print("La cantidad de muestras por especie es:")
print(counts_especies)


#  Troba l'espècie amb el valor màxim mitjana de la variable "sepal length (cm)"

#  numpy.argmax(a, axis=None, out=None, *, keepdims=<no value>)

sepal_length = datos[:, 0]

medias_sepal = [np.mean(sepal_length[clase == clase1])for clase1 in range(3)]


print(medias_sepal)

max_media_sepal_length = np.argmax(medias_sepal)

nombres_especies = ["setosa", "versicolor", "virginica"]

especie_max_media = nombres_especies[max_media_sepal_length]

print("La especie con el valor máximo medio de sepal length es: " ,especie_max_media)
```

```python
# Exercici 8: Troba l'espècie amb la diferència més gran entre 
# la longitud mitjana de sèpals i pètals

import numpy as np


# diferencia promedio entre longitud de sépalos y longitud de pétalos por especie
diferencias_promedio = []
for etiqueta in range(3):
    sepalos_por_especie = array_conjunto[(clase == etiqueta)][:, 0]  # Longitud de los sépalos para esta especie
    petalos_por_especie = array_conjunto[(clase == etiqueta)][:, 2]  # Longitud de los pétalos para esta especie
    diferencia_media = np.mean(sepalos_por_especie) - np.mean(petalos_por_especie)
    diferencias_promedio.append(diferencia_media)

# clase con la mayor diferencia promedio
etiqueta_max_diferencia = np.argmax(diferencias_promedio)
print("Etiqueta especie con diferencia promedio más alta: ", etiqueta_max_diferencia)


nombres_especies = ["setosa", "versicolor", "virginica"]
especie_max_diferencia = nombres_especies[etiqueta_max_diferencia]


print("Especie con mayor diferencia entre media de longitud sépalo y pétalo: ", especie_max_diferencia)
```
